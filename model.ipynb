```python
# Conversion function for Self Rating Python,R programming & Data science to weightage

def self_rating_weightage(self_rating):
    if self_rating == 1:
        return 3
    elif self_rating == 2:
        return 7
    elif self_rating == 3:
        return 10
    else:
        return 0
```

```python
# Conversion from self rating to weightage

data["tot_score"] = data['Python (out of 3)'].map(self_rating_weightage)
data["tot_score"] += data['R Programming (out of 3)'].map(self_rating_weightage)
data["tot_score"] += data['Data Science (out of 3)'].map(self_rating_weightage)
```

```python
data.head()
```

```python
# Extracting features out of Others skills

def other_skill_weightage(skills):
    weight = 0
    if "Machine Learning" in skills:
        weight +=3
    if "Deep Learning" in skills:
        weight +=3
    if "NLP" in skills:
        weight +=3
    if "Statistical Modeling" in skills :
        weight +=3
    if "AWS" in skills:
        weight +=3
    if "SQL" in skills:
        weight +=3
    if "NoSQL" in skills:
        weight +=3
    if "Excel" in skills:
        weight +=3
    return weight
```

```python
# Conversion from Other Skills rating to weightage

data['tot_score'] += data["Other skills"].map(other_skill_weightage)
```

```python
data.head()
```

```python
# Converting degree into PG or UG degree.

def degree_type(deg):
    deg_type=''
    if "B.Tech" in deg or 'Bachelor' in deg:
        deg_type="UG"
    elif "MBA" in deg or 'Master' in deg or "Integrated" in deg:
        deg_type="PG"
    elif "Post Graduate" in deg or "PG" in deg:
        deg_type="PG"
    else:
        deg_type="PG"
    return deg_type
```

```python
data["degree_type"] = data['Degree'].map(degree_type)
```

```python
data['degree_type'].value_counts()
```

```python
# Conversion function for Graduation Degree type & Year to weightage

def degree_year_weightage(deg_type, year): 
    weight = 0
    if deg_type == "UG" and year == '2020':
        weight = 10
    elif deg_type == "UG" and year == '2019':
        weight = 8
    elif deg_type == "UG" and year <= '2018':
        weight = 5
    elif deg_type == "PG" and year == '2020':
        weight = 7 
    elif deg_type == "PG" and year <= '2019':
        weight = 3
    else:
        weight = 0
    return weight
```

```python
# Conversion from Graduation and Year to weightage

data['tot_score'] += data.apply(lambda a: degree_year_weightage(a['degree_type'], a['Current Year Of Graduation']), axis=1)
```

```python
# Selecting application if the total score is greater than equal to 40 then shortlisted for 'Data Scientist' or '1'

def shortlisted(weight):
    if weight >= 40:
        return 1
    else:
        return 0
```

```python
data["shortlisted"] = data['tot_score'].map(shortlisted)
```

## 3. Model Building and Evaluation

Let's start building the model. The first step to model building is the usual test-train split. So let's perform that

```python
# Put all the feature variables in X

X = data_num.drop(['shortlisted'], 1)
X.head()
```

```python
# Put the target variable in y

y = data_num['shortlisted']

y.head()
```

```python
# Split the dataset into 70% train and 30% test

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

### Scaling

Now there are a few numeric variables present in the dataset which have different scales. So let's go ahead and scale these variables.

```python
# Import MinMax scaler

from sklearn.preprocessing import StandardScaler
```

```python
# Scale the three numeric features present in the dataset

scaler = StandardScaler()

# Create a list of the variables that you need to scale
varlist = ['Python (out of 3)', 'R Programming (out of 3)','Data Science (out of 3)', 'Machine Learning', 
           'Deep Learning', 'NLP', 'Statistical Modeling', 'AWS', 'SQL', 'NoSQL', 'Excel', 'Graduation',
           'degree_type_UG','degree_type_PG']

# Scale these variables using 'fit_transform'
X_train[varlist] = scaler.fit_transform(X_train[varlist])

X_train.head()
```

```python
# # Heatmap to see correlation between variables

# # Correlation matrix
# cor = data_num.corr()
# plt.figure(figsize=(16,8))

# # Heatmap
# sns.heatmap(cor, annot=True)
# plt.show()
```

```python
# Import 'LogisticRegression' and create a LogisticRegression object
logreg = LogisticRegression()
```

```python
# Fit a logistic Regression model on X_train

logreg.fit(X_train, y_train)
```

```python
# Use 'predict' to predict the probabilities on the train set

y_train_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_train_pred)
```
